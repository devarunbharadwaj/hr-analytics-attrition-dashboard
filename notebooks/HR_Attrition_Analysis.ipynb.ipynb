# Exploratory Data Analysis (EDA)

```python
attrition_percent = df['Attrition'].value_counts(normalize=True) * 100

attrition_percent.plot(kind='bar')

plt.title("Attrition Percentage")
plt.xlabel("Attrition")
plt.ylabel("Percentage")

plt.show()
```

```python
department_attrition = pd.crosstab(df['Department'], df['Attrition'])

department_attrition.plot(kind='bar')

plt.title("Department-wise Attrition")
plt.xlabel("Department")
plt.ylabel("Count")

plt.xticks(rotation=0)

plt.show()
```

```python
overtime_attrition = pd.crosstab(df['OverTime'], df['Attrition'])

overtime_attrition.plot(kind='bar')

plt.title("OverTime vs Attrition")
plt.xlabel("OverTime")
plt.ylabel("Count")

plt.xticks(rotation=0)

plt.show()
```

```python
overtime_attrition = pd.crosstab(df['OverTime'], df['Attrition'])

overtime_attrition.plot(kind='bar')

plt.title("OverTime vs Attrition")
plt.xlabel("OverTime")
plt.ylabel("Count")

plt.xticks(rotation=0)

plt.show()
```

```python
plt.hist(df['MonthlyIncome'], bins=20)

plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")

plt.show()
```

```python
job_satisfaction_attrition = pd.crosstab(
    df['JobSatisfaction'],
    df['Attrition']
)

job_satisfaction_attrition.plot(kind='bar')

plt.title("Job Satisfaction vs Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Count")

plt.xticks(rotation=0)

plt.show()
```

Observation:
Employees working overtime show noticeably higher attrition compared to employees not working overtime.

```python
age_attrition = pd.crosstab(df['Age'], df['Attrition'])

age_attrition.plot(kind='line')

plt.title("Age vs Attrition")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```
